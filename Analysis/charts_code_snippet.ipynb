```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Create a dataframe for the Low Crime Rate Race Demographic
race_data = {'state': ['Alabama','Alaska','Arizona','Arkansas','California','Colorado','Connecticut','Delaware'],
       'Black': [191,5,4869,105,137,16,10,20],
       'American Indian/Alaskan Native': [0,414,1317,59,0,15,0,15],
       'Asian': [9,60,745,23,2851,230,74,53],
       'Native Hawaiian / Pacific Islander': [0,0,156,0,0,12,0,0,],
       'Hispanic/Latino': [0,34,19871,286,457,1205,32,75],
       'White': [3882,1635,29974,4540,7226,8599,579,1711],
       'Other': [0,1,4228,0,20,305,4,9],
       'Two or More': [64,242,1503,77,711,276,27,16]}

race_merge_df = pd.DataFrame(race_data)

race_merge_df
```

```python
# Tell matplotlib that we will be making a bar chart
# Race is our y axis and State Name, listed vertically, is our x axis
# We apply align="edge" to ensure our bars line up with our tick marks
#x_axis = race_merge_df['state']
#x_axis = np.arange(len(race_merge_df['state'])
#plt.bar(x_axis, race_merge_df['state'], color='r', alpha=0.5, align="center")
race_merge_df.plot.bar()
```

```python
# Previous chart was in Pandas; create chart using Matplotlib
# First, set variables for each data element
x_axis = race_merge_df['state']
race_black = race_merge_df['Black']
plt.bar(x_axis, race_black, color='r', alpha=0.5, align="center")
```

```python
# Now, create a stacked bar chart to compare with bar chart
race_merge_df.plot.bar(stacked=True)
```

```python
# Now, create the stacked bar chart in Matplotlib
# Define the rest of the variables for each data element
# Add each variable to the bottom of the bar chart to stack the bars
race_native_american = race_merge_df['American Indian/Alaskan Native']
plt.bar(x_axis, race_black, bottom=race_native_american, color='b', alpha=0.5, align="center")

# Problem: the Matplotlib stacked bar chart can only have two items stacked
# because it can have only one "bottom" argument
# Switching back to pandas charts until we can figure this out in Matplotlib
```
