```python
# Dependencies
import numpy as np 
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
```

## Plot low crime vs high crime summary table for marital status

```python
maritalsummarytable = pd.read_csv("Data/marital_summary_table.csv")
```

```python
# Set x axis and tick locations
fig, ax = plt.subplots()
x_axis = np.arange(len(maritalsummarytable))
tick_locations = [value for value in x_axis]
# Create a list indicating where to write x labels and set figure size to adjust for space
#plt.figure(figsize=(8,3))
ax.bar(x_axis, maritalsummarytable["% Never Married"], color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, maritalsummarytable["Low/High"])
# Set x and y limits
plt.xlim(-0.50, len(x_axis)-.50)
plt.ylim(0, max(maritalsummarytable["% Never Married"])+3)
# Set a Title and labels
plt.title("Marital Status: Low Crime vs High Crime City")
plt.ylabel("Percentage of Never Married")
#Display value on top of bar
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height*1.02), ha='center')
# Save our graph and show the graph
plt.savefig("Results/NeverMarried.png")
plt.show()
```

```python
educationsummarytable=pd.read_csv("Data/education_summary_table.csv")
#Reconfigure dataframe for line chart
lcnone=educationsummarytable["% None"][0]
lchs=educationsummarytable["% High School"][0]
lcas=educationsummarytable["% Associates"][0]
lcba=educationsummarytable["% Bachelor"][0]
lcgr=educationsummarytable["% Graduate"][0]
hcnone=educationsummarytable["% None"][1]
hchs=educationsummarytable["% High School"][1]
hcas=educationsummarytable["% Associates"][1]
hcba=educationsummarytable["% Bachelor"][1]
hcgr=educationsummarytable["% Graduate"][1]
neweducation = {'Education':["None","High School","Associates","Bachelor","Graduate"],
                'Low Crime Cities': [lcnone,lchs,lcas,lcba,lcgr], 
                'High Crime Cities': [hcnone,hchs,hcas,hcba,hcgr]}
neweducation_df= pd.DataFrame(neweducation)
#plot line chart
x_axis_data = np.arange(len(neweducation_df["Education"]))
tick_locations = [value for value in x_axis_data]
low=neweducation_df["Low Crime Cities"]
high=neweducation_df["High Crime Cities"]
plt.plot(x_axis_data, low, label="Low Crime", marker='o')
plt.plot(x_axis_data, high, label="High Crime", marker='o')
plt.xticks(tick_locations, neweducation_df["Education"])
plt.legend(loc='best')
plt.title('Comparison of Education Level', fontsize=14)
plt.xlabel('Education Level', fontsize=14)
plt.ylabel('Percentage', fontsize=14)
plt.grid(True)

# Save our graph and show the graph
plt.savefig("Results/education.png")
plt.show()
```

```python
schoolsummarytable=pd.read_csv("Data/school_summary_table.csv")
#Reconfigure dataframe for line chart
lcn=schoolsummarytable["% Nursery-Pre-School"][0]
lck=schoolsummarytable["% Kindergarten"][0]
lc4=schoolsummarytable["% 1st-4th"][0]
lc8=schoolsummarytable["% 5th-8th"][0]
lc12=schoolsummarytable["% 9th-12th"][0]
hcn=schoolsummarytable["% Nursery-Pre-School"][1]
hck=schoolsummarytable["% Kindergarten"][1]
hc4=schoolsummarytable["% 1st-4th"][1]
hc8=schoolsummarytable["% 5th-8th"][1]
hc12=schoolsummarytable["% 9th-12th"][1]
newschool = {'School Enrollment':["Nur.-Pre-School","Kindergarten","1st-4th","5th-8th","9th-12th"],
                'Low Crime Cities': [lcn,lck,lc4,lc8,lc12], 
                'High Crime Cities': [hcn,hck,hc4,hc8,hc12]}
newschool_df= pd.DataFrame(newschool)
#plot line chart
x_axis_data = np.arange(len(newschool_df["School Enrollment"]))
tick_locations = [value for value in x_axis_data]
low=newschool_df["Low Crime Cities"]
high=newschool_df["High Crime Cities"]
plt.plot(x_axis_data, low, label="Low Crime", marker='o', color='r')
plt.plot(x_axis_data, high, label="High Crime", marker='o',color='g')
plt.xticks(tick_locations, newschool_df["School Enrollment"])
plt.legend(loc='best')
plt.title('Comparison of School Enrollment', fontsize=14)
plt.xlabel('School Enrollment Level', fontsize=14)
plt.ylabel('Percentage', fontsize=14)
plt.grid(True)

# Save our graph and show the graph
plt.savefig("Results/schoolenrollment.png")
plt.show()
```
