```python
# This code creates chartes from demographic datasets
# We will chart the following demographics against the Low Crime data and High Crime data
# The demographic data points are:
#     Income level
#     Population
#     Marital Status
#     Race
#     Education level
#     School Enrollment Status
#     Unemployment level
```

```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read Low Crime Income Level csv into a new dataframe
lowcrime_income_df = pd.read_csv('Low Crime/Data/incomecrime.csv')

# Display a scatter chart showing states on the x_axis and median income on the y-axis
# Use Matplotlib

# Set x axis and tick locations
x_axis = np.arange(len(lowcrime_income_df))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)
    
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.scatter(x_axis, lowcrime_income_df["Median Income"], color='b', alpha=0.75)
plt.xticks(tick_locations, lowcrime_income_df["State"], rotation="vertical")

# Add Title and Labels to chart
plt.title("Median Income for Lowest Crime Rate City Within Each State")
plt.ylabel("Median Income (in dollars)")

# Save Low Crime - Income scatter chart
plt.tight_layout()
plt.savefig("Low Crime/Results/lowcrime_income.png")
plt.show()
```

```python
# Read Low Crime Marital Status csv into a new dataframe
lowcrime_marital_df = pd.read_csv('Low Crime/Data/maritalstatuscrime.csv')
# Display a stacked bar chart showing states on the x_axis and 100 on on the y-axis
# The marital status types will make up the bar stacks
# Use Matplotlib horizontal bar chart showing discrete distributions

lowcrime_marital_df=lowcrime_marital_df.set_index("State")
fig, ax = plt.subplots(figsize=(12, 6))
x = lowcrime_marital_df.index
indexes = np.argsort(lowcrime_marital_df.values).T
heights = np.sort(lowcrime_marital_df.values).T
order = -1
bottoms = heights[::order].cumsum(axis=0)
bottoms = np.insert(bottoms, 0, np.zeros(len(bottoms[0])), axis=0)
mpp_colors = dict(zip(lowcrime_marital_df.columns, plt.rcParams['axes.prop_cycle'].by_key()['color']))
for btms, (idxs, vals) in enumerate(list(zip(indexes, heights))[::order]):
    mps = np.take(np.array(lowcrime_marital_df.columns), idxs)
    ax.bar(x, height=vals, bottom=bottoms[btms], color=[mpp_colors[m] for m in mps])
ax.set_ylim(bottom=0, top=100)
ax.set_title("Marital Status for Lowest Crime Rate City Within Each State")
ax.set_ylabel("Proportion of Marital Status")
plt.xticks(rotation=90)
plt.legend((np.take(np.array(lowcrime_marital_df.columns), 
                    np.argsort(lowcrime_marital_df.values)[0]))[::order], loc='upper right')
#plot and save
plt.savefig("Low Crime/Results/lowcrime_marital.png")
plt.show()
```
